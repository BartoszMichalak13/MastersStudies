```python
import pandas as pd
import matplotlib.pyplot as plt

# Load results
df = pd.read_csv("results.csv")

# Get all unique m values
m_values = sorted(df["m"].unique())
ncols = 3
nrows = (len(m_values) + ncols - 1) // ncols

# Create subplots
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15, 4 * nrows))
axes = axes.flatten()

for i, m in enumerate(m_values):
    subdf = df[df["m"] == m]
    ax = axes[i]
    ax.scatter(subdf["n"], subdf["n_hat_over_n"], s=1)
    ax.set_title(f"m = {m}")
    ax.set_xlabel("n")
    ax.set_ylabel("n̂ / n")
    ax.grid(True)

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
