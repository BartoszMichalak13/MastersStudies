```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv("src/results.csv")
# sem1/AlgorytmyOnline/lista5/src/results.csv
# Ensure types are correct
df["D"] = df["D"].astype(int)
df["p"] = df["p"].astype(float)

# Unique p-values
ps = sorted(df["p"].unique())

# Plot Average Cost vs D for each p
plt.figure(figsize=(10, 6))
for p in ps:
    subset = df[df["p"] == p]
    plt.plot(subset["D"], subset["AverageCost"], marker='o', label=f"p={p}")
plt.xlabel("D (copy cost)")
plt.ylabel("Average Cost")
plt.title("Average Cost vs D for various p")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("average_cost_vs_D.png")
plt.show()

# Plot Avg Max Copies vs D for each p
plt.figure(figsize=(10, 6))
for p in ps:
    subset = df[df["p"] == p]
    plt.plot(subset["D"], subset["AvgMaxCopies"], marker='o', label=f"p={p}")
plt.xlabel("D (copy cost)")
plt.ylabel("Average Max Copies")
plt.title("Average Max Copies vs D for various p")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("average_max_copies_vs_D.png")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data (replace 'results.csv' with your actual output filename)
df = pd.read_csv('src/trace.csv')

# Group by D and P, and compute average over all requests
grouped = df.groupby(['D', 'p']).agg({
    'AverageCost': 'mean',
    'AvgMaxCopies': 'mean'
}).reset_index()

# Get all distinct D values
ds = sorted(df['D'].unique())

# Plot settings
colors = ['tab:blue', 'tab:orange']
markers = ['o', 's']

for d in ds:
    subset = grouped[grouped['D'] == d]

    fig, ax1 = plt.subplots(figsize=(8, 5))
    ax2 = ax1.twinx()

    ax1.plot(subset['p'], subset['AverageCost'], color=colors[0], marker=markers[0], label='Average Cost')
    ax2.plot(subset['p'], subset['AvgMaxCopies'], color=colors[1], marker=markers[1], label='Avg Max Copies')

    ax1.set_xlabel('p (Write Probability)')
    ax1.set_ylabel('Average Cost', color=colors[0])
    ax2.set_ylabel('Avg Max Copies', color=colors[1])

    ax1.tick_params(axis='y', labelcolor=colors[0])
    ax2.tick_params(axis='y', labelcolor=colors[1])

    plt.title(f'Results for D = {d}')
    fig.tight_layout()
    plt.savefig(f'plot_D{d}.png')
    plt.close()

print("Plots saved: plot_D*.png")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('src/trace.csv')

# Ensure correct types
df['D'] = df['D'].astype(int)
df['p'] = df['p'].astype(float)

# Unique D and p values
ds = sorted(df['D'].unique())
ps = sorted(df['p'].unique())

# Plot for each D
for d in ds:
    df_d = df[df['D'] == d]

    # --------- Plot AverageCost ---------
    plt.figure(figsize=(10, 6))
    for p in ps:
        df_dp = df_d[df_d['p'] == p]
        grouped = df_dp.groupby("Request").mean(numeric_only=True).reset_index()
        plt.plot(grouped['Request'], grouped['AverageCost'], label=f'p={p}')

    plt.title(f'Average Cost vs Request Index (D = {d})')
    plt.xlabel('Request')
    plt.ylabel('Average Cost')
    plt.legend()
    plt.tight_layout()
    plt.savefig(f'avg_cost_D{d}.png')
    plt.close()

    # --------- Plot AvgMaxCopies ---------
    plt.figure(figsize=(10, 6))
    for p in ps:
        df_dp = df_d[df_d['p'] == p]
        grouped = df_dp.groupby("Request").mean(numeric_only=True).reset_index()
        plt.plot(grouped['Request'], grouped['AvgMaxCopies'], label=f'p={p}')

    plt.title(f'Avg Max Copies vs Request Index (D = {d})')
    plt.xlabel('Request')
    plt.ylabel('Avg Max Copies')
    plt.legend()
    plt.tight_layout()
    plt.savefig(f'avg_copies_D{d}.png')
    plt.close()

print("Plots saved: avg_cost_D*.png and avg_copies_D*.png")
```

```python
window = 100  # Adjust this for more or less smoothing

for d in ds:
  df_d = df[df['D'] == d]
  plt.figure(figsize=(10, 6))

  for p in ps:
    df_dp = df_d[df_d['p'] == p]
    grouped = df_dp.groupby("Request").mean(numeric_only=True).reset_index()
    grouped['SmoothedCost'] = grouped['AverageCost'].rolling(window, min_periods=1).mean()
    plt.plot(grouped['Request'], grouped['SmoothedCost'], label=f'p={p}')

  plt.title(f'Smoothed Average Cost vs Request Index (D = {d})')
  plt.xlabel('Request')
  plt.ylabel('Smoothed Average Cost')
  plt.legend()
  plt.tight_layout()
  plt.savefig(f'smoothed_avg_cost_D{d}.png')
  plt.close()
```
