```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Załaduj dane
df = pd.read_csv("results.csv")

# Ustawienia stylu wykresów
sns.set(style="whitegrid")
palette = sns.color_palette("tab10")

# Lista algorytmów w ustalonej kolejności
algorithms = ["FIFO", "FWF", "LRU", "LFU", "RAND", "RMA"]

# Dla każdego rozkładu — osobny wykres
for dist in df["distribution"].unique():
    plt.figure(figsize=(10, 6))
    subset = df[df["distribution"] == dist]

    for alg in algorithms:
        data = subset[subset["algorithm"] == alg]
        grouped = data.groupby("n")["cost"].mean().reset_index()
        plt.plot(grouped["n"], grouped["cost"], marker='o', label=alg)

    plt.title(f"Średni koszt żądania – rozkład: {dist}")
    plt.xlabel("Liczba stron (n)")
    plt.ylabel("Średni koszt")
    plt.legend()
    plt.tight_layout()

    filename = f"plot_{dist}.png"
    plt.savefig(filename)
    print(f"Zapisano wykres: {filename}")
    plt.show()
```
