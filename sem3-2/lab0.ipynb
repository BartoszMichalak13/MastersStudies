# Zad 1

Wygeneruj próbę losową o dł. n z rozkładu: 
 - wykładniczego o parametrze \lambda > 0 
 - Pareta z pramatetrem \alpha > 0, \lambda > 0

Przypomnienie:
 - X ~ Exp(\lambda); F_x(t) = 1 - e^{-t\lambda}, t \geq 0
 - E X = 1/\lambda

 - X ~ Par(\alpha, \lambda) gdy F_x(t) = 1 - (\frac{\lambda}{\lambda + t})^alpha, t > 0

Uwaga:
Rozkład Pareto często jest nazywany Lomax. Często pod tą nazwą jest rozkłąd o F_x(t) = 1 - (\frac{X_o}{t})^\alpha, t\geq X_o

Ponadto wygeneruj wykresy x_n od n

Hint: można X =d= F^{-1}_x(U), U ~ U(0,1), U uzyskujemy z rand

```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def exp(lambda_param, n):
  results = []
  for i in range(n):
    U = random.uniform(0, 1)
    x = -1 / lambda_param * np.log(U)
    results.append(x)
  return results
```

```python
def pareto(alpha_param, lambda_param, n):
  results = []
  for i in range(n):
    U = random.uniform(0, 1)
    x = lambda_param / (U ** (1 / alpha_param))
    results.append(x)
  return results
```

## Verify results

```python
X = exp(0.5, 10000)
print(X)

expected_value = sum(X) / len(X)
print("Expected value:", expected_value, "Theoretical value:", 1 / 0.5)
```

```python
plt.hist(X, bins=50)
plt.title("Histogram of Exponential Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
X_05 = pareto(0.5, 1, 10000)
print(X_05)
expected_value = sum(X_05) / len(X_05)
print("Expected value:", expected_value, "Theoretical value:", float('inf'))
# Expected value of Pareto distribution is infinite for alpha <= 1, so we cannot calculate it here.
# for alpha > 1, the expected value is given by lambda * alpha / (alpha - 1).

X_1 = pareto(1, 1, 10000)
print(X_1)
expected_value = sum(X_1) / len(X_1)
print("Expected value:", expected_value, "Theoretical value:", float('inf')) # This will be infinite, so we cannot calculate it.  

X_2 = pareto(2, 1, 10000)
print(X_2)
expected_value = sum(X_2) / len(X_2)
print("Expected value:", expected_value, "Theoretical value:", 1 * 2 / (2 - 1))

X_3 = pareto(3, 1, 10000)
print(X_3)
expected_value = sum(X_3) / len(X_3)
print("Expected value:", expected_value, "Theoretical value:", 1 * 3 / (3 - 1))
```

```python
plt.hist(X_05, bins=50)
plt.title("Histogram of Pareto Distribution (alpha=0.5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(X_1, bins=50)
plt.title("Histogram of Pareto Distribution (alpha=1)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

# Zad 2

Wygenerować próbę Y-ków

Y_k = 1/N \sum^N_{i=1} X_i, k = 1,2,1000; N - dł. próby X-ów

dla X-ów z rozkładu Pareta z \lambda = 1,
\alpha = 1/2, 3/2, 3

```python
def gen_Ys(alpha_param, lambda_param, n, k):
  Ys = []
  for i in range(k):
    X = pareto(alpha_param, lambda_param, n)
    Y = 1 / n * sum(X)
    Ys.append(Y)
  return Ys
```

```python
Y_0_5 = gen_Ys(0.5, 1, 10000, 1000)
Y_1_5 = gen_Ys(1.5, 1, 10000, 1000)
Y_3 = gen_Ys(3.0, 1, 10000, 1000)

print("Y_0_5:", Y_0_5)
print("Y_1_5:", Y_1_5)
print("Y_3:", Y_3)
```

```python
plt.hist(Y_0_5, bins=50)
plt.title("Histogram of Y for Pareto Distribution (alpha=0.5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(Y_1_5, bins=50)
plt.title("Histogram of Y for Pareto Distribution (alpha=1.5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(Y_3, bins=50)
plt.title("Histogram of Y for Pareto Distribution (alpha=3)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
